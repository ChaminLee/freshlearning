```python
import tensorflow as tf
tf.set_random_seed(777)

x_train = [1,2,3]
y_train = [1,2,3]
```

```python
W = tf.Variable(tf.random_normal([1]),name = 'weight')
b = tf.Variable(tf.random_normal([1]),name = 'bias')
```

```python
hypo = W*x_train + b
```

```python
cost = tf.reduce_mean(tf.square(hypo-y_train))
```

```python
Optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = Optimizer.minimize(cost)
```

```python
sess = tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost),sess.run(W),sess.run(b))
```

```python
W = tf.Variable(tf.random_normal([1]),name='weight')
b = tf.Variable(tf.random_normal([1]),name='bias')
```

```python
X = tf.placeholder(tf.float32, shape = [None])
Y = tf.placeholder(tf.float32, shape = [None])
```

```python
hypo = W*X + b
cost = tf.reduce_mean(tf.square(hypo-Y))
```

```python
Optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = Optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cost_val,W_val,b_val,_ = \
        sess.run([cost,W,b,train], feed_dict = {X : [1,2,3],Y : [1,2,3]})
    if step % 20 == 0:
        print(step,cost_val,W_val,b_val)
        
```

```python
W = tf.Variable([.3],tf.float32)
b = tf.Variable([-.3],tf.float32)

x = tf.placeholder(tf.float32,shape=[None])
y = tf.placeholder(tf.float32,shape=[None])
```

```python
hypo = W*x + b
```

```python
loss = tf.reduce_mean(tf.square(hypo-y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(loss)
```

```python
x_train = [1,2,3,4]
y_train = [0,-1,-2,-3]
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```

```python
if step in range(1000):
    sess.run(train,{x : x_train, y : y_train})
curr_W,curr_b,curr_loss = sess.run([W,b,loss], feed_dict={x:x_train,y:y_train})
print("W:%s b:%s loss:%s" % (curr_W,curr_b,curr_loss))

```
