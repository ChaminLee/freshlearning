```python
import numpy as np
```

```python
xy = np.loadtxt("data-01-test-score.csv",delimiter=',', dtype=np.float32)
```

```python
x_data = xy[:,0:-1]
y_data = xy[:,[-1]]
```

```python
y_data
```

```python
print(x_data.shape,x_data,len(x_data))
```

```python
print(y_data.shape,y_data)
```

```python
import tensorflow as tf
```

```python
tf.set_random_seed(777)
```

```python
X = tf.placeholder(tf.float32,shape=[None,3])
Y = tf.placeholder(tf.float32,shape=[None,1])
```

```python
W = tf.Variable(tf.random.normal([3,1]),name="weight")
b = tf.Variable(tf.random.normal([1]),name="bias")
```

```python
hypothesis = tf.matmul(X,W) + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
for i in range(2001):
    cost_val, hy_val,_ = sess.run([cost,hypothesis,train],feed_dict={X : x_data, Y : y_data})
    if i % 10 == 0:
        print(i,"cost is:",cost_val,"hy is",hy_val)
```

```python
filename_queue = tf.train.string_input_producer(["data-01-test-score.csv"],shuffle=False,name="filename_queue")
```

```python
reader = tf.TextLineReader()
```

```python
key, value = reader.read(filename_queue)
```

```python
record_defaults = [[0.],[0.],[0.],[0.]]
```

```python
xy = tf.decode_csv(value, record_defaults=record_defaults)
```

```python
train_x_batch, train_y_batch = \
tf.train.batch([xy[0:-1], xy[-1:]], batch_size=10)
```

```python
X = tf.placeholder(tf.float32, shape=[None,3])
Y = tf.placeholder(tf.float32, shape=[None,1])
```

```python
W = tf.Variable(tf.random_normal([3,1]),name="weight")
b = tf.Variable(tf.random_normal([1]),name="weight")
```

```python
hypo = tf.matmul(X,W) + b
```

```python
cost = tf.reduce_mean(tf.square(hypo - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
coord = tf.train.Coordinator()
```

```python
threads = tf.train.start_queue_runners(sess=sess,coord=coord)
```

```python
for i in range(2001):
    x_batch, y_batch = sess.run([train_x_batch, train_y_batch])
    cost_val, hy_val,_ = sess.run([cost,hypo,train],feed_dict = {X : x_batch, Y : y_batch})
    if i % 10 == 0:
        print(i,"cost :", cost_val,"hy :",hy_val)
```

```python
coord.request_stop()
```

```python
coord.join(threads)
```
