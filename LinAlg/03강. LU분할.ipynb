\mathbf{E_{31}}=\begin{bmatrix}

\end{bmatrix}
\end{equation*}


\begin{equation*}
\mathbf{E_{32}}=\begin{bmatrix}

\end{bmatrix}
\end{equation*}

\mathbf{E_{21}A}=\begin{bmatrix}

\end{equation*}

위의 결과는 Gaussian elimination의 결과이다. ($l_{21}$ 이용한 것)

\begin{equation*}
\mathbf{E_{21}A}=\begin{bmatrix}

### 1.5 Triangular Factors & Row Exchanges

\begin{equation*}
\mathbf{Ax}=\begin{bmatrix}
2 & 1 & 1 \\\
4 & -6 & 0 \\\
-2 & 7 & 2
\end{bmatrix}
\begin{bmatrix}
u\\\
v\\\
w
\end{bmatrix}=
\begin{bmatrix}
5\\\
-2\\\
9
\end{bmatrix}=
b
\end{equation*}
        
    
    * step 1 : row2 - $l_{21}$ * row1      => row2'
    * step 2 : row3 - $l_{31}$ * row1      => row3'
    * step 3 : row3' - $l_{32}$ * row2'
    
식 (2)와 (3)의 $u$를 먼저 제거해주고, 그 다음 step 3에서 $v$를 제거하고 있다. 

여기서 참고하자면 $ E_{ab}$ 라는 것은 $b$의 식을 이용해서 $a$식을 바꾸겠다는 것이다. 

결국, 결과는 "Upper Triangular Matrix"(U)로 나온다. 

\begin{equation*}
\mathbf{E_{32}E_{31}E_{21}A} = \begin{bmatrix}
2 & 1 & 1 \\\
0 & -8 & -2 \\\
0 & 0 & 1
\end{bmatrix}
\end{equation*}


오른쪽 위에만 숫자가 있는 Upper Triangular(U) 모양으로 나왔다. ( Pivot ≠ 0 )

Gauss Elimination을 하면 아래와 같이 정리된다.

\begin{equation*}
\begin{cases}
2u + v + w = 5\\\
-8v -2w = -12\\\
w = 2
\end{cases}
\end{equation*}

이는 다시 행렬로 표현하면 다음과 같다.

\begin{equation*}
\mathbf{Ux}=\begin{bmatrix}
2 & 1 & 1\\\
0 & -8 & -2\\\
0 &0 & 1
\end{bmatrix}
\begin{bmatrix}
u\\\
v\\\
w
\end{bmatrix}=
\begin{bmatrix}
5\\\
-12\\\
2
\end{bmatrix}=
c
\end{equation*}

$x$는 $Ux=c$에서 back-substitution하면 구할 수 있다. ( w,v,u 순으로 구함 )

Matrix에 Inverse를 해주면


$A = E_{21}^{-1}E_{31}^{-1}E_{32}^{-1}U$

위와 같은 모양으로 나온다. 

$E_{21}^{-1}E_{31}^{-1}E_{32}^{-1}$는 ( 3x3 역행렬을 통해 직접 구하면 나온다 )

\begin{equation*}
\mathbf{E_{21}^{-1}E_{31}^{-1}E_{32}^{-1}}=\begin{bmatrix}
1 & 0 & 0\\\
l_{21}& 1 & 0\\\
l_{31} & l_{32} & 1
\end{bmatrix}
\end{equation*}
(사진)

\begin{equation*}
\begin{bmatrix}
1 & 0 & 0\\\
2& 1 & 0\\\
0 &0 & 1
\end{bmatrix}
\begin{bmatrix}
1 & 0 & 0\\\
0& 1 & 0\\\
-1 &0 & 1
\end{bmatrix}
\begin{bmatrix}
1 & 0 & 0\\\
0& 1 & 0\\\
0 & -1 & 1
\end{bmatrix}=
\begin{bmatrix}
1 & 0 & 0\\\
\color{Red}2& 1 & 0\\\
\color{Red}-\color{Red}1 &\color{Red}-\color{Red}1 & 1
\end{bmatrix}=
L
\end{equation*}

$A = LU$  (L은 Lower Triangular Matrix를 의미한다/ 그래서 LU분할!) 

대각선 아래있는 성분들은 곱셈에 사용이 된다 $ l_{21},l_{31},l_{32}$
Elimination의 과정을 잘 기록되어있다.   ( 앞에서는 연산을 위해 다 -처리가 되어 있었음 )


### Triangular Factorization(descomposition = 인수분해 ) 

$A=LU$를 row changes없이 만들기!

$U$의 대각선 성분이 Pivot이 된다


$Ax=b$, first $Lc=b$, then $Ux=c$

(1) Factor : $A$에서 $L$과 $U$를 찾는다.
(2) Solve : $L$과 $U$에서 Solution $x$를 찾는다. 


* $U$는 $DU$로도 나타낼 수 있다.  (D = Diagnoal Matrix : Pivots의 matrix)


\begin{equation*}
\mathbf{U}=\begin{bmatrix}
d_1 & a & b & c\\\
0 & d_2 & d & e\\\
0 & 0 & d_3 &  f \\\
0 & 0 & 0 & d_4
\end{bmatrix}
\end{equation*}
일 때, 다음과 같이 나타낼 수 있다. 

\begin{equation*}
\mathbf{U}=\begin{bmatrix}
d_1 & 0 & 0 & 0\\\
0 & d_2 & 0 & 0\\\
0 & 0 & d_3 &  0 \\\
0 & 0 & 0 & d_4
\end{bmatrix}
\begin{bmatrix}
1 & \frac{a}{d_1} & \frac{b}{d_1} & \frac{c}{d_1}\\\
0 & 1 & \frac{d}{d_1} & \frac{e}{d_1}\\\
0 & 0 & 1 &  \frac{f}{d_1} \\\
0 & 0 & 0 & 1
\end{bmatrix}=
DU
\end{equation*}
위와 같이 표현할 수 있는데, 여기서 D를 Diagnoal Matrix라고 부르는 것이다. 

따라서 $A = LU = LDU$로 둘 다 가능하다.
$D$는 거듭제곱을 수행할 때 유용하게 작용한다.  

\begin{equation*}
\mathbf{D^{n}}=\begin{bmatrix}
{d_1}^{n} & 0 & 0 & 0\\\
0 & {d_2}^{n} & 0 & 0\\\
0 & 0 & \ddots &  0 \\\
0 & 0 & 0 & {d_n}^{n}
\end{bmatrix}
\end{equation*}

* $LDU$ factorization 과 $LU$ factorization은 행렬 $A$에 대해 unique하다 ( 조합이 두 가지 이상 안나옴)


### Row Exchange and Permutation Matrices

* zero in the Pivot position


\begin{equation*}
\begin{bmatrix}
0 & 2 \\
3 & 4 
\end{bmatrix}
\begin{bmatrix}
u \\
v 
\end{bmatrix}=
\begin{bmatrix}
b_1 \\
b_2
\end{bmatrix}
\end{equation*}

=> Exchange Rows 의 결과는 다음과 같다. ($P_{21}$은 2행, 1행 순서바꾼다는 것)

\begin{equation*}
\begin{cases}
3u + 4v = b_2\\
2v = b_1
\end{cases}
\end{equation*}

* Permutation

\begin{equation*}
\mathbf{P} = \begin{bmatrix}
0 & 1 \\
1 & 0 
\end{bmatrix}
\end{equation*}


\begin{equation*}
\mathbf{PA} = \begin{bmatrix}
3 & 4 \\
0 & 2 
\end{bmatrix}=
\mathbf{Pb} = \begin{bmatrix}
b_2\\
b_1 
\end{bmatrix}
\end{equation*}

#### Permutation matrix $P$는 $I$와 같은 수의 행을 가져야 한다. 또한 어느 행,열을 골라도 항상 1개의 1이 있어야 한다.

\begin{equation*}
\mathbf{I} = \begin{bmatrix}
1 & 0 & 0 \\\
0 & 1 & 0 \\\
0 & 0 & 1
\end{bmatrix},
\mathbf{P_{21}} = \begin{bmatrix}
0 & 1 & 0 \\\
1 & 0 & 0 \\\
0 & 0 & 1
\end{bmatrix},
\mathbf{P_{31}} = \begin{bmatrix}
0 & 0 & 1 \\\
0 & 1 & 0 \\\
1 & 0 & 0
\end{bmatrix}
\end{equation*}

\begin{equation*}
\mathbf{P_{32}} = \begin{bmatrix}
1 & 0 & 0 \\\
0 & 0 & 1 \\\
0 & 1 & 0
\end{bmatrix},
\mathbf{P_{32}P_{21}} = \begin{bmatrix}
0 & 1 & 0 \\\
0 & 0 & 1 \\\
1 & 0 & 0
\end{bmatrix}≠
\mathbf{P_{21}P_{32}} = \begin{bmatrix}
0 & 0 & 1 \\\
1 & 0 & 0 \\\
0 & 1 & 0
\end{bmatrix}
\end{equation*}

$P$행렬간의 교환법칙은 성립하지 않는다.

그리고 $P^{-1} = P^T$ 이다. ( 원래대로 되돌리는 것이다 )

(1) Non - Singular cases면

$PA = LU$ 이고, 따라서 $A = P^TLU$ 가 성립한다. 

(2) Singular cases면, $P$가 존재하지 않고, Elimination에 실패한다.
