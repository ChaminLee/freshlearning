#### length

$||x^2|| = |x_1| + |x_2| + \cdots + |x_n^2|$

$= {x_1}^* x_1 + {x_2}^* x_2 + \cdots + {x_n}^* x_n$

$=x^{*T} x$


#### Inner product

$x^{*T}y (≠y^{*T}x)$

$=x^Hy (x^H = x^{*T}$ ( Hermitian ) ( conjugate transpose )

(1) Orthogonal $x$ and $y$ = $x^Hy = 0$

(2) $||x||^2 = x^Hx = |x_1|^2 + |x_2|^2 + \cdots + |x_n|^2$

(3) $ (AB)^H = B^HA^H $

(4) $(x^Hy)^H = y^Hx$

$x^Hy <-> y^Hx$ "conjugate"

### Hermitian Matrices

* Real components : $A^T = A$ => "Symmetric"

* Complex : $A^H = A$ => "Hermitian"

$a_{ij} = a{ji}^* (i≠j)$

$a_ij  (i=j)$   (=real)


### Symmetric matrices $\subset$ Hermitian matrices



## Properties of Hermitian or Symmetric Matrix

### (1) $X^HAX$ (quadratic form) is "real"

$R=A^HA$ =>

$X^HRX = X^HA^HAX$

$=(AX)^HAX$

$=||AX||^2 >0$

### (2) If $A^H = A$, every eigenvalue is real

$Ax = \lambda x$ for complex vecctor x

각 항에 $x^H$를 곱하면

$x^HAx = x^H\lambda x = \lambda x^Hx = \lambda||x||^2$

$\lambda = \frac{x^HAx}{||x||^2}$   

분자, 분모 모두 real(실수)라서 람다도 real 이다. 

### (3) Eigenvectors are orthogonal

$Ax_1 = \lambda_1x_1 , Ax_2 = \lambda_2x_2, (\lambda_1≠\lambda_2 and A^H=A)$

$(Ax_1)^Hx_2 = x_1^H A^H x_2 = x_1^H A x_2$

$= (\lambda_1x_1)^Hx_2 = x_1^H\lambda_1^Hx_2 = \lambda_1x_1^Hx_2$

$x_1^HAx_2 = x_1^H\lambda_2x_2 = \lambda_2x_1^Hx_2 = \lambda_1x_1^Hx_2$

$= (\lambda_2-\lambda_1)x_1^Hx_2=0 (\lambda_1≠\lambda_2)$

그래서 

$x_1^Hx_2=0$  or  $x_2^Hx_1 = 0$ 이어서 수직이게 되는 것이다.
