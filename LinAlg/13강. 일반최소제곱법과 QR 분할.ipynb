## Orthogonal Basis & Gram-Schmitt 

$q_1,q_2,\cdots,q_n$ are orthogonal

* $q_i^Tq_j = 0 (i≠j)$
* $ q_i^Tq_j  = 1 (i=j)$

$Q = [q_1,q_2,\cdots,q_n]$

$Q^TQ = I$ => Left-inverse

=> $Q^{-1} = Q^T$

* Rotation matrix 

\begin{equation*}
\begin{bmatrix}
\cos \theta & -\sin \theta \\
\sin \theta & \cos \theta
\end{bmatrix}
\end{equation*}

* Permutation matrix

\begin{equation*}
\begin{bmatrix}
0 & 1 & 0 \\
0 & 0 & 1 \\
1 & 0 & 0
\end{bmatrix}
\end{equation*}

->  $ P^{-1} = P^T$


* Geometrically, an orthogonal $Q$ is the product of a rotation and a reflection

* Projection reduces the length of a vector,
    But orthonormal matrix preserves angles and lengths
    
$||Qx||^2 = x^TQ^TQx = ||x||^2 $ -> length conservation

$$(Q^TQ = I)$$

$(Qx)^T(Qy) = x^TQ^TQy = x^Ty$ -> inner product or angle conservation

* For any vector $b$

$b = x_1q_1 + x_2q_2 + \cdots + x_nq_n$ => $Qx=b$

$q_i^Tb=x_i$

$b = (q_i^Tb)q_1 + \cdots + (q_n^Tb)q_n$

$b = Qx$ <=> $x=Q^Tb = Q^{-1}b$


* 1-D projection onto a line $a$

= $\frac{a^Tb}{a^Ta}a$

for $q_i, \frac{q_i^Tb}{q_i^Tq_i}q_i = (q_i^Tb)q_i = x_iq_i$

-> $x_i$ : projection of $q_i$ of $b$


* $Q^T = Q^{-1}$ => $QQ^T=I$, for square cases
    The rows are also orthonormal
    
#### Rectangular matrix with Orthonormal columns

for m < n ; least square cases

$Qx=b$

$Q^TQ\hat{x}=Q^Tb$  : normal equation for least squares

$\hat{x} = Q^Tb$ : ($Q^TQ=I$)

$p = Q\hat{x}$

$p = Q(Q^TQ)^{-1}Q^Tb = QQ^Tb$

$Q^T$ 는 Left-inverse이다. 

m>n일 때는  $QQ^T$가 $I$로 나오지 않음.


### Gram-Schmitt Orthogonalization

-> find the orthonormal basis given independent vectors

사진

$q_1 = \frac{a}{||a||}$

$\frac{q_1^Tb}{q_1^Tq_1}q_1$

$B = b-(q_1^Tb)q_1 \bot q_1 $

이는 다음과 같다.

$\frac{b-(q_1^Tb)q_1}{||b-(q_1^Tb)q_1||}=q_2$

$b=(q_1^Tb)q_1+(q_2^Tb)q_2$
 
사진

$C = c - [(q_1^Tb)q_1+(q_2^Tb)q_2]=(q_3^Tc)q_3$

$C = \sum_{i=1}^3(q_i^Tc)q_i$

아래는 공식화 한 것이다.

$a_j = \sum_{i=1}^j(q_i^Ta_j)q_i$


### Factorization $A=QR$


\begin{equation*}
\begin{bmatrix}
\\
a_1 & a_2 & \cdots & a_n \\
\\
\end{bmatrix} = 
\begin{bmatrix}
(q_1^Ta_1)q_1 & (q_1^Ta_2)q_1 & (q_1^Ta_3)q_1 & \cdots \\\
0 & +(q_2^Ta_2)q_2 & + (q_2^Ta_3)q_2 & \cdots \\\
0 & 0 &+(q_3^Ta_3)q_3 & \sum_{i=1}^n(q_i^Ta_n)q_i \\\
\\\
\end{bmatrix}
\end{equation*}


* linear combination of each column vector

\begin{equation*}
\begin{bmatrix}
\\
q_1 & q_2 & \cdots & q_n \\
\\
\end{bmatrix} = 
\begin{bmatrix}
(q_1^Ta_1) & (q_1^Ta_2) & (q_1^Ta_3) & \cdots & q_1^Ta_n \\\
0 & (q_2^Ta_2) & + (q_2^Ta_3) & \cdots & \vdots\\\
0 & 0 &(q_3^Ta_3) & & \vdots \\\
\vdots & \vdots & \vdots & \ddots &  \vdots \\\
0 & 0 & 0 & \cdots & q_n^Ta_n
\end{bmatrix}
\end{equation*}


Upper triangular matrix 모양이다

이를 $A=QR$ 이라고 해서 QR분해라고한다.
