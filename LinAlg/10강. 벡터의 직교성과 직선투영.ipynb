## Orthogonality ( 직교성 )

### Orthogonal vectors and subspaces

* orthogonality -> independent basis, easy calculation
* coordinate axes are orthogonal
* the fundamental subspaces meet at right angles

* length of vector : $||x||^2 = \sum_{i=1}^n x_i^2 = x^Tx$

* for a right angle 

사선을 $||x||^2+||y||^2 = ||x-y||^2$으로 표현할 수 있다. 

위 식을 풀어보면 다음과 같다

$$x^Tx+y^Ty = (x-y)^T(x-y)$$

$$= x^Tx-y^Tx-x^Ty+y^Ty $$

$$x^Ty=y^Tx=0$$  

=> $x^Ty$ ( 결국 내적이다 )

$x^Ty=0$은 "수직"이라는 것을 말한다. ( 벡터 내적의 결과 )

-> Linearly independent ( Linear combination 했을 때, 모든 계수가 0이어야 영벡터가 될 떄 )

* $x^Ty=0$ : for orthogonal ( 직교한다 )
* $x^Ty<0$ : for angle > 90 degree
* $x^Ty>0$ : for angle < 90 degree


* If nonzero vectors $V_1,V_2,\cdots,V_k$ are mutually orthogonal
-> then the vectors are independent

$C_1V_1 +C_2V_2+\cdots+C_kV_k=0$

for any $V_i, V_i^T$ = $C_i||V_i||^2=0$

C를 쉽게 G.E없이 구할 수 있다. 


## Orthogonal subspaces


-> Every vector in one subspace must be orthogonal to every vector in the subspace

$v \in \mathbb{V}, w \in \mathbb{W}$ 일 때, (in $\mathbb{R}^n$) 

$v^Tw=0$ for all of $v$ and $w$

$v,w$는 직교한다는 것이다. 

여기서 Dim(v)+Dim(w)=n 이라면 " Orthogonal Complete Subspace " 라고 한다. 

$V \in R^n$, the space of all vectors orthogonal to $V$ is called the 
" Orthogonal Complete Subspace " of $V$


* Dim of Vector Space = number of independent vectors to span the Vector Space = rank of A (=r)


* row space $\bot$ null space ( in $\mathbb{R}^n$)

( r + ( n-r ) = n(col) )

* column space $\bot$ left nullspace ( in $\mathbb{R}^m$)

( r + ( m-r ) = m(row) )

## Cosines and Projections onto lines 

사진

벡터 b가 a에 수선의 발을 내리는 것을 " projection b onto a " 라고 한다.

$(b-\hat{p}a) \bot a$ => $a^T(b-\hat{p}a)=0$


$\hat{p}=\frac{a^Tb}{a^Ta}$

사진

제 2 코사인 법칙에 따라서 다음과 같이 전개된다.

$$||b-a||^2=||a||^2+||b||^2-2||a||||b||\cos\theta$$

$$(b-a)^T(b-a) = a^Ta+b^Tb-2||a||||b||\cos\theta$$

$$a^Tb=||a||||b||\cos\theta$$

결국 내적을 통해 Projection과 같은 결과를 얻을 수 있다. 

수선의 발은 $\hat{p}a$ 이다. 이를 수정해 보면 

$\hat{p}=\frac{a^Tb}{a^Ta}a$

$=\hat{p}=\frac{aa^T}{a^Ta}b$ 가 된다. 

$$\mathbb{P}=\frac{aa^T}{a^Ta}$$ 

이를 Projection Matrix라고 한다. 

p를 구하기 위해서는 $\hat{p}=\frac{aa^T}{a^Ta}$에 투영하고자 하는 vector "b"를 곱하면 된다.

* $\mathbb{P}^T$ = $\mathbb{P}$

* $\mathbb{P}^2=\mathbb{P}$

$$\mathbb{P}=\frac{aa^T}{a^Ta}$$ 

이는 다음과 같은 행렬로 표현할 수 있다. 

\begin{equation*}
\begin{bmatrix}
C  \\
S
\end{bmatrix}
\begin{bmatrix}
C & S
\end{bmatrix} = 
\begin{bmatrix}
C^2 & CS \\
CS & S^2
\end{bmatrix}
\end{equation*}
