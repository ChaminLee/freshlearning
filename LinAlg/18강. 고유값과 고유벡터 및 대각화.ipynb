## Eigenvalues & Eigenvectors

$Ax = \lambda x$

$\lambda$ : eigenvalue , $x$ : eigenvector

=> row exchanges change the eigenvalue $\lambda$

=> Use determinant

$det(A-\lambda I) = 0 $  -> Singular 

$\mathbb{N}(A-\lambda I)$ 

=> how to find null-space ( eigen vector ) 

=> Reduced Row Echelon form ( pivot variables / free variables )

Ex)

\begin{equation*}
\mathbf{A}=\begin{bmatrix}
4 & -5 \\
2 & -3 
\end{bmatrix}
\end{equation*}

\begin{equation*}
\mathbf{A-\lambda I}=\begin{bmatrix}
4-\lambda & -5 \\
2 & -3-\lambda 
\end{bmatrix}
\end{equation*}

$det(A-\lambda I) = (4-\lambda)(-3-\lambda) + 10 $

$=> \lambda^2-\lambda-2=0$

## Diagonalization of a Matrix

=> The eigenvectors diagonalization a matrix

For nxn matrix $A$, $A$ ha n linearly independent eigenvectors ($e_i$)

넘어가기 전에 Decomposition에 3가지 방법이 있다. 

(1) $A=LU=LDU$

(2) $A=QR$  (=Gram-schmitt)

(3) $A=S\Lambda S^{-1}$

\begin{equation*}
\mathbf{S}=\begin{bmatrix}
\\\
e_1 & e_2 & \cdots & e_n \\\\
\end{bmatrix}
\end{equation*}


\begin{equation*}
\mathbf{S^{-1}AS=\Lambda}=\begin{bmatrix}
\lambda_1 & & & &\\\
& \lambda_2 & & &\\\
& & \lambda_3 & & \\\
& & & \ddots & \\\\
& & & & \lambda_n
\end{bmatrix}
\end{equation*}

-----------------

\begin{equation*}
\begin{bmatrix}
\\\
\lambda e_1 &\lambda e_2 & \cdots &\lambda e_n \\\\
\end{bmatrix}=
\begin{bmatrix}
\\\
e_1 & e_2 & \cdots & e_n \\\\
\end{bmatrix}
\begin{bmatrix}
\lambda_1 & & & &\\\
& \lambda_2 & & &\\\
& & \lambda_3 & & \\\
& & & \ddots & \\\\
& & & & \lambda_n
\end{bmatrix}
\end{equation*}

으로 표현이 된다!

### Assume that $e_2$ is dependent on $e_1$

$e_2 = ce_1 $ ( independent -> c=0 )

양변에 $A$를 곱해준다

$Ae_2 = cAe_1$

$=> \lambda_2e_2 = c\lambda_1 e_1$

기존 식에 양변에 $\lambda_2$를 곱해준다

$\lambda_2 e_2 = c\lambda_2 e_1$

결국에는 

$c(\lambda_2-\lambda_1)e_1$이 성립해야한다.

하지만 $e_1$은 영행렬이 아니고, $\lambda_2≠\lambda_1$ 이다.

따라서 $c=0$이어한다,

그렇다면 $e_1, e_2$는 independent하다. 가설이 틀렸다는 것 이다. 

일반화 해봐도 똑같은 결과로 도출된다. 

따라서 $e_n$들은 linearly independent 하다.

## Remark 

If $\lambda_1, \lambda_2, \cdots, \lambda_n$ are distinct(다르다면), than n eigenvectors are independent

(1) $ S = te$ => eigenvector

(2) $S$ is not Unique, $A_n$ eigenvector can be mutiplied by a constant

(3) The order of eigenvalues and eigenvectors in $S$ and $\Lambda$ are the same

$e_n , \lambda_p$ n,p는 순서에 맞게 쌍을 이루어야 한다. 

(4) Not all matrices have n linearly independent eigenvectors, so not all matrices are diagonalizable
