### Vector Space

= Set of vectors in $\mathbb{R}^n$

(1) Closed under addition ( 덧셈이 가능하다 )

$$ V_1 \in \mathbb{V}, V_2 \in \mathbb{V} $$

$$ V_1 + V_2 \in \mathbb{V} $$

(2) Closed under scalar multiplication ( 스칼라곱이 가능하다 )

$$ V \in \mathbb{V}, c \in \mathbb{R} $$

$$ cV \in \mathbb{V} $$

(3) $\mathbb{V}$ includes the zerovector( 원점 )

$$ \mathbb{V} = \left\{0\right\} $$


* Column space of $A$

= Set of all linear combinations of column vectors in $A$

$$\left\{x | x = \sum_{i=1}^n c_ia_i\right\}$$

$$ A = [a_1, a_2,\cdots, a_n ] $$

$$Ax = b $$

\begin{equation*}
\mathbf{x} = \begin{bmatrix}
x_1 \\\
x_2 \\\
\vdots \\\
x_n
\end{bmatrix}
\end{equation*}

$x_1a_1 + x_2a_2 + \cdots + x_na_n = b $

if, $b \in \mathbb{C}(A)$  -> there are solutions ( 1개 이상 )

if, $b \notin \mathbb{C}(A) $  -> no solution


* Span : 벡터들로 만들 수 있는 모든 Linear Combinations 집합 
    Whole space is constructed by linear combinations
    
### Null space of A ( $N(A)$ )

= Set of vectors such that $Ax=0$

There are infinite solutions  

They form a vector space - null space of A

* The null space of a matrix A consists of all vectors x such that $Ax=0$, 
    $N(A)$ -> subspace
    
    
(i) Closed under addition

$Ax = 0$ and $ Ax'=0$ -> $A(x+x')=0$, closed under addition

(ii) Closed under scalar multiplication

$Ax = 0$ -> $ Acx = c0 = 0 $, closed under scalar multiplication (for any C)


ex) 
(1)

\begin{equation*}
\mathbf{A} = \begin{bmatrix}
1 & 0 \\
5 & 4\\
2 & 4 
\end{bmatrix}
\begin{bmatrix}
u \\
v\\
\end{bmatrix}=
\begin{bmatrix}
0 \\
0 \\
0
\end{bmatrix}
\end{equation*}

-> homogeneous equation!  $ Ax = 0 $

$N(A) = (0,0)$

(2)
\begin{equation*}
\mathbf{B} = \begin{bmatrix}
1 & 0 & 1 \\
5 & 4 & 9\\
2 & 4 & 6
\end{bmatrix}
\begin{bmatrix}
u \\
v \\
w
\end{bmatrix}=
\begin{bmatrix}
0 \\
0 \\
0
\end{bmatrix}
\end{equation*}

-> same $\mathbb{C}(A), \mathbb{C}(B)$

--> $N(B)$ is in the line  (c,c,-c) (1,1,-1)

we will find $N(A), \mathbb{C}(A)$

## Solving $Ax=0$(null space) & $Ax=b$

미지수 개수(unknowns) > 식의 개수(eqns)   

=> 해가 아예 없거나(평행), 무수히 많다. 

* For an invertible matrix $A$, the null space $N(A)$ contains only $x=0$, 
    The column space is the whole space
    
* When the null space contains more than the zero vector column space contains less than all vectors

* $b$ complete solution $Ax_p=b$ and $Ax_n=0$ -> $A(x_p+x_n) = b $


ex) 
\begin{equation*}
\begin{bmatrix}
1 & 1 \\
2 & 2\\
\end{bmatrix}
\begin{bmatrix}
y \\
z \\
\end{bmatrix}=
\begin{bmatrix}
b_1 \\
b_2 
\end{bmatrix}
\end{equation*}

(i) $b_2 ≠ 2b_1$ 

-> No-Solution

(ii) $b_2 = 2b_1$ 

-> infinitely many solutions

$x_p$ = (1,1)   

=> \begin{equation*}
\begin{bmatrix}
1 & 1 \\
2 & 2\\
\end{bmatrix}
\begin{bmatrix}
y \\
z \\
\end{bmatrix}=
\begin{bmatrix}
2 \\
4 
\end{bmatrix}
\end{equation*}

$x_p$가 1,1이면 b_1과 b_2는 2,4가 된다. 

$x_n$ = (-1,1) or (-C, C)

\begin{equation*}
\mathbf{x_p+x_n} = \begin{bmatrix}
1 \\
1\\
\end{bmatrix}+
\mathbf{C}\begin{bmatrix}
-1 \\
1 \\
\end{bmatrix}=
\begin{bmatrix}
1-C \\
1+C 
\end{bmatrix}
\end{equation*}

$x=x_p+x_n$에 대한 해결책들을 line으로 나타낸다.

-> the simplest matrix that elimination can give

\mathbf{A} = \begin{bmatrix}

\mathbf{A} = \begin{bmatrix}

3번식 - 2* 2번식

### Pivot Variables and Free Variables

for solving $Ax = 0$, the Pivots are " important "

\begin{equation*}
\mathbf{Ax} = \begin{bmatrix}
1 & \color{red}3 & 0 & \color{red}-\color{red}1 \\
0 & \color{red}0 & 1 & \color{red}1\\
0 & 0 & 0 & 0
\end{bmatrix}
\begin{bmatrix}
u\\
v\\
w\\
y
\end{bmatrix}=
\begin{bmatrix}
0\\
0\\
0
\end{bmatrix}
\end{equation*}


빨간색으로 표시된 성분들은 " Special solution ( reverse sign ) " 이다.

Pivots는 첫 번째, 세 번째 열에 위치해 있다. 

* Pivot Variables with pivot columns = $u, w$
* Free Variables without pivot columns = $v, y$

solve = $ u + 3v - y = 0$ and $ w + y = 0$

$$u = -3v + y$$

$$w = -y $$


\begin{equation*}
\mathbf{x} = \begin{bmatrix}
u\\
v\\
w\\
y
\end{bmatrix}=
\begin{bmatrix}
-3v+y\\
v\\
-y \\
y
\end{bmatrix}=
\mathbf{v}\begin{bmatrix}
-3\\
1\\
0\\
0
\end{bmatrix}+
\mathbf{y}\begin{bmatrix}
1\\
0\\
-1\\
1
\end{bmatrix}
\end{equation*}


* Special Solution : (-3,1,0,0) , (1,0,-1,1)  <= (v=1, y=0) and (v=0, y=1)

=> Null space contains all combinations of special solutions using free variables

v,y 와 같은 개수가 null space의 diemension이 됨. 모양새가 4차원 같지만 사실은 2차원이다.

(미지수 = 4 , 표현 가능 벡터 = 2)
