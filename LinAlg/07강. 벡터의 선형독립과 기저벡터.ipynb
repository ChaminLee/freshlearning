## Linear Independence ( 선형독립 )
## Basis(vectors) , Dimension

### Linear independent

$ C_1V_1 + C_2V_2 + \cdots + C_nV_n = 0$( 0은 영벡터)

only $C_1=C_2=\cdots=C_n=0$ 만이 solution

* 스칼라 값이 0이 될때만, 그 벡터들을 선형 결합해 만들 수 있을 때 $V_1,V_2$를 Linear independent 하다고 말한다. 

ex) 

\begin{equation*}
\mathbf{C_1}\begin{bmatrix}
1\\\
0\\\
0
\end{bmatrix}+
\
\mathbf{C_2}\begin{bmatrix}
0\\\
1\\\
0
\end{bmatrix}+
\mathbf{C_3}\begin{bmatrix}
0\\\
0\\\
1
\end{bmatrix}=
\begin{bmatrix}
0\\\
0\\\
0
\end{bmatrix}
\end{equation*}

* $C_1, C_2, C_3=0 $ 이어야 0이된다. 이럴 때 세 벡터는 linear independent하다고 말한다. 


* If $G.E$ of $A$ generates 'm' non-zero rows, 

$U$ 형태에서 Pivot을 포함해 적어도 0이 아닌 수가 하나 있어야함.

"m"개의 independent한 vectors가 column에 존재한다 (in $A$)

다 0이 아닌게 한 개씩 존재하면 linear independent 한것

non-zero 행만큼이 = independent m의 개수
