#### Relation between eigenvalues and special solutions ( $e^{At}$ of differential equation)


\begin{equation*}
\mathbf{\frac{du(t)}{dt}}=\begin{bmatrix}
a & b\\
c & d
\end{bmatrix}
\begin{bmatrix}
u(t) \\
\end{bmatrix}
\end{equation*}


\begin{equation*}
\begin{bmatrix}
x'(t)\\
y'(t)
\end{bmatrix}=
\begin{cases}
ax(t) + by(t) \\
cx(t) + dy(t)
\end{cases}
\end{equation*}

먼저 $x(t)$를 구해보자


$x(t) = \frac{1}{c}(y'(t)-dy(t))$ 이다.


이를 이용해서 $x'(t)$를 구해보면 서로 다른 두 식이 나온다. 이를 같다고 두고 전개하면 다음과 같다. 



$\frac{1}{c}(y''(t)-dy'(t) = \frac{a}{c}(y'(t)-dy(t)) + by(t)$

이를 잘 정리하면 다음과 같다. 

$y''(t) - (a+d)y'(t) + (ad-bc)y(t) = 0$

이는 Characteristic equation으로, $\lambda^2 - (a+d)\lambda + (ad-bc) = 0$으로 나타낼 수 있다. 


따라서 $y(t) = c_1e^{\lambda_1t} + c_2 e^{\lambda_2t}$


* by eigenvalue problem of A

\begin{equation*}
\mathbf{det}\begin{bmatrix}
a-\lambda & b\\
c & d-\lambda
\end{bmatrix}= (a-\lambda)(d-\lambda) - bc = 0
\end{equation*}

이를 전개하면 

$\lambda^2 - (a+d)\lambda + (ad-bc) = 0$으로 위에서 구한 결과와 같다!


$\frac{d}{dt}u(t) = Au(t)$ -> Find eigenvalue A to get $e^{\lambda t}$

$u(t) = c_1e^{\lambda_1t} + c_2 e^{\lambda_2t}$




--------

\begin{equation*}
\mathbf{u(0) = c_1x_1 + c_2x_2}=\begin{bmatrix}
x_1 & x_2\\
\end{bmatrix}
\begin{bmatrix}
c_1\\
c_2
\end{bmatrix}= Sc , c=S^{-1}u(0)
\end{equation*}


\begin{equation*}
\mathbf{u(t)}=\begin{bmatrix}
x_1 & x_2\\
\end{bmatrix}
\begin{bmatrix}
e^{\lambda_1 t} & 0\\
0 & e^{\lambda_2 t}
\end{bmatrix}
\begin{bmatrix}
x_1 & x_2\\
\end{bmatrix}^{-1}
u(0)=
Se^{\Lambda t }S^{-1}u(0)
\end{equation*}


Solution of $\frac{d}{dt}u(t) = Au(t)$ 

#### Taylor Series

$e^x = 1 + x + \frac{x^2}{2!} + \cdots$

x를 At로 / 1을 $I$로 대체!

$e^At = I + At + \frac{{At}^2}{2!} + \cdots$


$\frac{d}{dt}(e^{At}) = Ae^{At}$

$e^At = I + At + \frac{{At}^2}{2!} + \cdots$

$ = I + S\Lambda S^{-1} t + \frac{S\Lambda^2 S^{-1} t^2}{2!} + \cdots$

$ = S( I + \Lambda t + \frac{(\Lambda t)^2}{2!} + \cdots )S^{-1}$

$ = Se^{\Lambda t} S^{-1}$
