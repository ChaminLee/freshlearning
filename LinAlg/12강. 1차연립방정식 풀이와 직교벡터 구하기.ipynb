# 제 12강. 1차연립방정식 풀이와 직교벡터 구하기

이전 내용의 중요한 내용을 복습해 보자면 다음과 같다.

#### (1) Square System ( $A_{m \times n } : m=n$ )

정사각행렬로 기본적인 $G.E.$를 통해 Unique한 Solution을 구할 수 있다. 

$Ax=b$ 형태에서 $A$의 역행렬을 이용하여

$x=A^{-1}b$로 $x$를 구할 수 있다. 

#### (2) Under constrained system ( m (식) < n (미지수) )

가로로 긴 직사각형 모양을 가진 행렬 $A$를 말한다. 

미지수가 더 많기 때문에 보통은 " Infinitely many solution "이다. 

$ x = x_n + x_p$으로 $x$를 나타낼 수 있다. 

이는 무수히 많은 해집합이 " nullspace "와 평행하게 존재한다는 것을 말한다. ( $x_n$ : nullspace , $x_p$ : 평행이동 )

$N(A)$ = {$x_n | Ax_n=0$}

* How to find the solution 
    -> $G.E$ -> " Reduced Row Echelon "
    -> Special sol / particular sol
    -> Pivot var / Free var 들의 선형결합
    
#### (3) Over constrained system ( m (식) > n (미지수) )

세로로 긴 직사각형 모양을 가진 행렬 $A$를 말한다. 

일반적으로 " No solution " 이라고 말한다. 

거리라도 최소화하기 위해 L2 norm을 이용한다. 

$$||Ax-b||^2$$ 

이것이 Least Square이다 ( = Projection ) 

#### If given independent vectors

$a_1,a_2,\cdots,a_n$

(1) find the orthogonal basis vector   ( 직교 벡터 구하기 ) 

(2) Gram Schmitt Orthogonalization

*  Gram Schmitt Orthogonalization

사진

(1) $q_1 = \frac{a}{||a||}$ a의 크기로 나누기 ( normalization )

(2) project $b$ on to $q_1$

$\frac{q_1^Tb}{q_1^Tq_1}q_1$

$b-(q_1^Tb)q_1 \bot q_1 $

이는 다음과 같다.

$\frac{b-(q_1^Tb)q_1}{||b-(q_1^Tb)q_1||}=q_2$

$b=(q_1^Tb)q_1+(q_2^Tb)q_2$

두 개 벡터 Projection의 합이다. 


*********

규칙적으로 나타내면 아래와 같다

(1) $q_1 = \frac{a}{||a||}$

(2) $a_j - \sum_{i=1}^{j-1}(q_i^Ta_j)q_i$

이는 $A_j$와 같다.

(3) normalization

$q_j = \frac{A_j}{||A_j||}$

길이가 거의 0에 가까워짐 -> 버리고 다음으로 넘어간다

$a_j = \sum_{i=1}^j(q_i^Ta_j)q_i$
