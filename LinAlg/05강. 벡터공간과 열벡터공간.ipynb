## Vector Spaces and Subspaces

* Space : Set closed under addition & scalar multiplication
               원소갖는 집합, 덧셈 닫힘 / scala 곱 닫힘 ( 닫힘 = 포함, 가능 )

for any vectors  $X,Y \in \mathbb{R}^n$

for any scalar $ C \in \mathbb{R}^n$

$ X, Y \in \mathbb{V} $

\begin{equation*}
\begin{cases}
X+Y \in \mathbb{V}\\
CX \in \mathbb{V}
\end{cases}
\end{equation*}

$C_1X + C_2Y \in \mathbb{V}$    (V는 벡터공간)

### A real vector space $\mathbb{R}^n$(n-dimmensional)

(1) $ x + y = y + x $

(2) $ x + ( y + z ) = ( x + y ) + z

(3) There is a unique "Zero vector", such that $ x + 0 = 0 + x = x$

(4) For each $x$, there is a unique vector $-x$, such that $ -x + x = 0 $ ( 역원:unique하다)

(5) $ 1• x = x $

(6) $(C_1C_2)x = C_1(C_2x)

(7) $ C(x+y) = Cx + Cy $

(8) $ (C_1 + C_2)x = C_1x + C_2x $


\begin{equation*}
\begin{cases}
x, y   :  Vector\\
C,C_1,C_2   :  Scalar
\end{cases}
\end{equation*}

* $x$ and $y$ should be cloased under addition and scalar multiplication    => " Space " 

ex) $m$ x $n$ matrix $\in \mathbb{R}^{mn}$

2x2 matrix 이면  $\in \mathbb{R}^{mn}$ 


* Subspace : non-empty subset that satisfies the requirements for a vector space 
      
(i) $ x + y \in subspace$ ,  for $\forall x, y \in subspace $

(ii) $ cx \in subspace$ , for $\forall x \in subspace, \forall c$  


=> $x$ and $y$ should be closed under vector addition and scalar multiplications in the subspace

* different from subset 

ex) (i) $\mathbb{R}^3$ vectors on a plane(평면) which passes through the origin(원점)

(ii) the origin => smallest subspace

(iii) $\mathbb{R}^2$ vectors on a line which passes through the origin. 

(iv) lower triangular matrices

$ c_1L_1 + c_2L_2 \ in \mathbb{L} $

$ L_{nxn} \subset A_{nxn} \in \mathbb{R}^{n^2}$ 

* The Column Space of A 

- Column space : contains all linear combinations of the columns of matrix A

$ Ax = b $ (mxn)   => 

\begin{equation*}
\begin{bmatrix}
1 & 1 & \cdots & 1 \\\
a_1 & a_2 & \cdots & a_n\\\
1 & 1 & \cdots & 1
\end{bmatrix}
\begin{bmatrix}
x_1\\\
x_2 \\\
\vdots \\\
x_n
\end{bmatrix}=
\begin{bmatrix}
b_1\\\
b_2 \\\
\vdots \\\
b_n
\end{bmatrix}
\end{equation*}


$x_1a_1 + x_2a_2 + \cdots + x_na_n = b $

-> linear combinations of column vectors equals $b$


* $Ax = b$는 Solvable할 수 있다. 만약 $b$가 $A$의 Column의 결합으로 표현될 수 있다면 가능하다. 그러면 $b$는 column space에 있는 것이다. 

\begin{equation*}
\mathbf{u}\begin{bmatrix}
1 \\\
5\\\
2
\end{bmatrix}+
\mathbf{v}\begin{bmatrix}
0\\\
4\\\
4
\end{bmatrix}=
\begin{bmatrix}
b_1\\\
b_2 \\\
b_3
\end{bmatrix}
\end{equation*}

평면 위에 존재한다면 -> 해가 존재하고

평면 밖에 있다면 -> 해가 존재하지 않는다.


* Column Space 

if $b_1 ,b_2 \in \mathbb{C}(A)$ A라는 column space의 원소이다.

$Ax_1 = b_1 $

$Ax_2=b_2$

$b_1+b_2 = b $

$ Ax_1 + Ax_2 = A(x_1+x_2) = b $

* Column space $\mathbb{C}(A)$

(i) if $b$ and $b'$ lie in the column space -> $A(x+x') = b + b'

-> 덧셈에 대해 닫혀있다.(가능하다)

(ii) if $b$ lies in the column space -> $A(cx) = cb$

-> 스칼라 곱에 대해 닫혀있다. 

ex) (i) for any non-singular 5x5 matrix -> $\mathbb{R}^5$ (column) space

-> (5 pivots in Gauss Elimination )

(ii) Zero matrix -> 0 - dimensional space
