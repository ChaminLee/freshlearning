## Least Square Problem

$Ax=b$ => Non-Homogeneus equation

$A^TAx=A^Tb$

$x=(ATA)^{-1}A^Tb$ ( = Pseudo inverse )

min$||Ax-b||^2$

반면에 만약에,

$Ax  0 $  => Homogeneus equation

min$||Ax||^2$

=> $(Ax)^T(Ax)$ => $X^TA^TAX$

($||x||^2=1$로 제한을 둠 )

따라서

$J(x,\lambda) = ||Ax||^2 + \lambda(1-||x||^2)$

여기서 $\lambda$를 "Lagrange multiplier"이라고 한다.

그리고 각각 변수에 대해 편미분을 한다

$\frac{\partial J} {\partial x} = 0, \frac{\partial J} {\partial \lambda} = 0$

결국 다음과 같은 식이 도출될 것이다. 

$A^TAX=\lambda x$
