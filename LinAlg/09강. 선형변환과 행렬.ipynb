## Linear Transformations

$Ax=b$

$b$ is a linear combination of column vectors of $A$ with coefficient in $x$


$x$ is transformed(mapped) into $b$ by $A$

#### Stretching ( Extending or Contracting )
\begin{equation*}
\mathbf{A}=\begin{bmatrix}
C & 0 \\
0 & C
\end{bmatrix}
\end{equation*}

#### 90 degree rotation

\begin{equation*}
\mathbf{A}=\begin{bmatrix}
0 & -1 \\
-1 & 0
\end{bmatrix}
\end{equation*}

#### Reflection by $x_2=x_1$   ( $y=x$대칭)

\begin{equation*}
\mathbf{A}=\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
\end{equation*}

#### Projection into $x$ ( 수선의 발 )

\begin{equation*}
\mathbf{A}=\begin{bmatrix}
1 & 0 \\
0 & 0
\end{bmatrix}
\end{equation*}


$A_{m \times n}$ 일 떼, $\mathbb{R}^n$->$\mathbb{R}^m$으로 만드는 과정이다. ( Transform )

#### Ex) differentiation $\frac{d}{dt}$ of polynomial ( 다항식 )

$x(t) = a_0+a_1t + a_2t^2+\cdots+a_nt^n$


$y(t) = \frac{d}{dt}x(t) = b_0+b_1t+b_2t^2 + \cdots + b_{n-1}t^{n-1}$

$n+1$차원을 $n$차원으로 Transform하는 것이 미분이다. 

($a_0$부터 $a_n$까지 $n+1$개 / $b_0$부터 $b_{n-1}$까지 $n$개)


#### Ex) integration

 $\mathbb{R}^{n+1}$->$\mathbb{R}^{n+2}$의 과정이다. 
 
 null space는 무조건 "0"이다. 
 
#### Ex) multiplication by polynomial

다항식 간의 곱인데 2차원과 n차원간의 곱이라면 이는

 $\mathbb{R}^{n+1}$->$\mathbb{R}^{n+3}$으로 나타난다.
 

* If given $Ax$ for every basis vectors $X$

$Ax_1 = a_1$, $Ax_2 = a_2$, $\cdots$, $Ax_n = a_n$

then we can find any transform results in the vector space without $A$

#### Polynomial case : differentiation & integration

(1) first find a basis ( = most fundamental )

(2) then determine transformation (matrix) on the basis 

#### Ex) degree 3 polynomial $\mathbb{P_3}$

(1) basis for $\mathbb{P_3}$

\begin{equation*}
\mathbf{A}=\begin{bmatrix}
1 \\
t \\
t^2\\
t^3
\end{bmatrix}
\end{equation*}


$P_1 = 1, P_2= t, P_3=t^2, P_4 = t^3$


$\frac{d}{dt}A : AP_1 = 0, AP_2=1, AP_3=2t, AP_4 = 3t^2$


a vector form of each basis polynomial

$P_1 = (1,0,0,0), P_2=(0,1,0,0), P_3=(0,0,1,0),P_4=(0,0,0,1)$

\begin{equation*}
\mathbf{A_{diff}}=\begin{bmatrix}
0 & 1 & 0 & 0 \\
0 & 0 & 2 & 0 \\
0 & 0 & 0 & 3\\
0 & 0 & 0 & 0
\end{bmatrix}
\end{equation*}


* Free variable = $P_1$
* Pivot variable = $P_2, P_3,P_4$

null space는 상수항인 $P_1$만 0벡터로 만들어 주면 가능하게 된다.
(special solution)

* integration도 똑같은 과정으로 하면 된다. 


적분하고 미분하면 그대로가 되듯이, 

$A_{diff}A_{int}=I$이다. 

이는 differentiation 이 integration의 left-inverse라고 한다.

사진

$Q_\theta ^{2} = Q_{2\theta}$

$Q_\theta Q_\phi  = Q_{\theta+\phi}$

위는 삼각함수를 이용한 것이다. 


(2) Projection

$\theta$ line에 projection( 수선의 발)을 하는 것이다. 

\begin{equation*}
