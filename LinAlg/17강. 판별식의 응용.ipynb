### (4) Cofactor

Focus on one row !

$det(A) = \sum_{j=1}^n a_{ij}C_{ij}$

$C$는 $ a_{ij}$의 행, 열을 제외한 minor matrix이다. 

$C_{ij} = (-1)^{i+j}detM_{ij}$

따라서 다음과 같다. 

$det(A) = \sum_{j=1}^n a_{ij}(-1)^{i+j}detM_{ij}$


# Determinant Applications 

### (1) 주로 Cofactor와 관련이 있다. 

$\mathbb{C} $= Cofactor matrix

$\mathbb{C} $는 $C_{ij}$로 이루어진 Matrix이다. 


$$A^{-1} = \frac{\mathbb{C}^T}{det(A)}$$

위 식을 정리하면 다음과 같다. ( det(A)넘기고 양변에 A곱하고 ) 

$$(det(A))I = A\mathbb{C}^T$$

$A$ 는 $a_{ij}$로 이루어져있다.

두 개의 행이 같으면 $det$는 0이다. 

대각성분 말고 다른 성분을 계산할 때, 예를 들어, 위의 두 개의 행이 같아지는 모습이 보인다. 

이와 같을 때는 $det$가 0이다.

대각 성분의 경우 앞선 Cofactor의 조합과 같기 때문에 $det(A)$로 보여진다.


### (2) $Ax=b$ , $x=A^{-1}b$

Cramer's Rule

$x_i = \frac{det(B_i)}{det(A)}$

$B$는 I번째에 해당하는 column vector를 벡터 b를 넣어준다. (나머지는 다 a성분 ) 


$$A^{-1}b = \frac{\mathbb{C}^Tb}{det(A)}$$

(1)에 따라 여기에 b를 곱해준것이 된다. 

ex) 

\begin{equation*}
\begin{cases}
x+y=4 \\
2x-y=-1\\
\end{cases}
\end{equation*}

\begin{equation*}
\begin{bmatrix}
1 & 1 \\
2 & -1\\
\end{bmatrix}
\begin{bmatrix}
x\\
y\\
\end{bmatrix}=
\begin{bmatrix}
4\\
-1\\
\end{bmatrix}
\end{equation*}

$det(A) = -3 $

$det(B_1)$ 은 첫 번째 column vector를 $b$로 대체해서 구한다

\begin{equation*}
\begin{bmatrix}
4 & 1\\
-1 & 1\\
\end{bmatrix}=
-3
\end{equation*}

$det(B_2)$ 는 두 번째 column vector를 $b$로 대체해서 구한다

\begin{equation*}
\begin{bmatrix}
1 & 4\\
2 & -1\\
\end{bmatrix}=
-9
\end{equation*}

$ x = \frac{-3}{-3}=1$

$ y = \frac{-9}{-3}=3$ 


### (3) Volume of Box 

벡터 두 개 $a_1, a_2$에 대하여 $det$를 구하면 평행사변형이 나온다. 

$a_3$가 추가되어 3차원이 되었을 때, $det$의 값은 육면체의 부피가 된다. 

$$def(A) = Volume $$

* for right-angled box

$|a|=l_1 , |a_2|=l_2$

$a_i \bot a_j$ 서로 수직이다. 


$A^TA$는 서로 수직인 것 끼리 곱해지기 때문에 거리고 표현가능하다. 

n 번째 끼리만 길이에 대한 제곱으로 나오고 ( 대각 행렬 ) 나머지는 0 이다. 

(대각 성분 : $||a_1||^2, ||a_2||^2, ||a_3||^2,\cdots, ||a_n||^2$)

$det(A^TA)$는 $det(A)^2$으로 나올 것이다. 

그렇다면 성분들은 길이의 제곱인 $l_1^2,l_2^2,\cdots,l_n^2 $ 으로 나온다. 


* for non-right angled box ( 수직이 아닐 때 )

Gram schmitt방법을 Projection하여 높이를 구해줘야한다.
