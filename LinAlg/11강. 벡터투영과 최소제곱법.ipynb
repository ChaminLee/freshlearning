# 제 11강. 벡터투영과 최소제곱법


$(b-\hat{x}a) \bot a$ = $(b-\hat{x}A) \bot A$

=> $A^T(b-\hat{x}A)=0$

$A^Tb - A^TA\hat{x}=0$

$A^Tb = A^TA\hat{x}$

$\hat{x} = (A^TA)^{-1}A^Tb $

$\mathbb{P} = A\hat{x}$

$= A(A^TA)^{-1}A^Tb$

투영의 개념과 최소제곱법의 개념이 한 번에 다 들어가있다. 


## Projections and Least Squares

* For more equations than the unknowns, there is usually no solution.

=> Overconstrained cases ( mxn matrix ( m > n ) )



* How to find an optimal Solution ? 

=> minimizing errors $||Ax-b||^2 = E^2$ ( 거리의 개념으로 다가가기 )

#### Least square solution of 1 unknown -> projection onto the line a


* Orthogonality of a and e

-> The error vector $e$ connecting $b$ to $p$ must be perpendicular(수직) to $a$

$a^T(b-\hat{x}a)$

#### Least square problems with several variables

to project b onto a subspace ( m x n : A ( m > n ) )

$E = ||Ax=b||$

사진

$\hat{x}$ : solution of least squares to minimize $E$

$e=(b-A\hat{x}) \bot space$

#### (1) Since column space is perpendicular to left nullspace, $b-A\hat{x}$ is m the left nullspace

$A^T(b-A\hat{x})=0$ or $A^TA\hat{x}=A^Tb$

#### (2) The error vector must be orthogonal to each column vector

$a_i^T(b-A\hat{x})=0$

\begin{equation*}
\begin{bmatrix}
a_i^T \\
\vdots \\
a_n^T
\end{bmatrix}
\begin{bmatrix}
b-A\hat{x}
\end{bmatrix}=0
\end{equation*}

$a_i^T$들을 $A^T$로 표현할 수 있다.

$A^T[b-A\hat{x}] = A^Tb = A^TA\hat{x}$



* Normal equation = $A^TA\hat{x}=A^Tb$ ( 역행렬 있는지 모를 때 )

* Best estimate $\hat{x}$ for invertibility of $A^TA$ 
    => $\hat{x}=(A^TA)^{-1}A^Tb$

* Projection => $p = A\hat{x} = A(A^TA)^{-1}A^Tb$



#### Remark 1. 

if $b$ is in $\mathbb{C}(A)$ -> $b=Ax$

The projection of $b$ is still $b$

vector안에 존재해서 가까운게 자기 자신임

$ p =A(A^TA)^{-1}A^Tb = A(A^TA)^{-1}A^TAx = Ax = b $

#### Remark 2.

If $b$ is perpendicular to every column

$A^Tb=0$ -> $b$ in left nullspace

$p=A(A^TA)^{-1}A^Tb=0$

-> projected onto zero

#### Remark 3.

$A$ is square & invertible => $p=b$

$ p =A(A^TA)^{-1}A^Tb = A(A^{-1})(A^T)^{-1}A^Tb=b $
