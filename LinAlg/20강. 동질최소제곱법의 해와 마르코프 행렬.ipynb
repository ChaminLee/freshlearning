# Markov Matrix

상태가 변화하는 것을 의미한다. 이는

State transition in Probability 라고 한다. 

$U_{k+1} = AU_k$

$U_k = Au_0$

$\lambda = 1, |\lambda| < 1$

이것이 "Markov Process의 특징"이다.


\begin{equation*}
\mathbf{U_k}=\begin{bmatrix}
e_1 & e_2\\
\end{bmatrix}
\begin{bmatrix}
\lambda_1^k & 0\\\
0& \lambda_2^k 
\end{bmatrix}
\begin{bmatrix}
C_1  \\
C_2
\end{bmatrix}
\end{equation*}




\begin{equation*}
\begin{bmatrix}
y_k\\
z_k
\end{bmatrix}= C_1 \lambda_1^k e_1 + C_2 \lambda_2^k e_2
\end{equation*}


$\lim\limits_{k \to 0} U_k$ -> Converge ( 수렴한다 )

이를 "Stable State"라고 한다. 

$U_s = AU_s$



#### Difference equation 


\begin{equation*}

\begin{equation*}

## A Markov Matrix 

### (1) $a_ij ≥ 0 $, each column adds up to 1 ( 열 다 합치면 1 ) 

### (2) $\lambda = 1$ is an eigenvalue

### (3) eigenvector $x_1 ≥ 0$, steady state $ Ax_1 = x_1$

### (4) Markov Matrix has an eigenvalue = 1, the other eigenvalues $|\lambda < 1$

### (5) $A^ku_0$ -> $cx_1 -> u_∞$ ; steady state

### (6) Markov Matrix -> transition matrix ( for random process )
