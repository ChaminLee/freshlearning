### 2. $det(A)$ equals the volume of a box in $\mathbb{R}^n$ space

Ex) Jacobian

### 3. $det(A) = $ + or - Product of the Pivots

### 4. Cramer's rule for $Ax=b$

$x_i = \frac{det(A_i)}{det(A)}$

Ex)  
\begin{equation*}
\begin{cases}
x+y=3 \\
2x-y=3
\end{cases}
\end{equation*}

위의 식을 행렬로 바꾸면

\begin{equation*}
\begin{bmatrix}
1 & 1 \\
2 & -1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}=
\begin{bmatrix}
3 \\
3
\end{bmatrix}
\end{equation*}

위와 같이 표현할 수 있다.


\begin{equation*}
\mathbf{A_x}=\begin{bmatrix}
3 & 1 \\
3 & -1
\end{bmatrix}
\end{equation*}

$x$를 구할 때, 좌측 열을 모두 3을 바꿔준다. 

$$x = \frac{det(A_x)}{det(A)} = \frac{-6}{-3} = 2$$

y도 같은 방법으로 구한다.

이번엔 우측 열을 모두 3으로 바꿔준다. 

\begin{equation*}
\mathbf{A_y}=\begin{bmatrix}
1 & 3 \\
2 & 3
\end{bmatrix}
\end{equation*}

$$y = \frac{det(A_y)}{det(A)} = \frac{-3}{-3} = 1$$


## Properties Of Determinant

determinant by $2 \times 2$ matrix

\begin{equation*}
\mathbf{det}=\begin{bmatrix}
a & b \\
c & d
\end{bmatrix} = ad-bc
\end{equation*}

### Basic Properties

#### 1. $det(I)=1$

#### 2. The determinant changes sign(부호) when two rows are exchanged

#### 3. The determinant depends linearly on the first row

\begin{equation*}
\begin{bmatrix}
a+a' & b+b' \\
c & d
\end{bmatrix}=
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}+
\begin{bmatrix}
a' & b' \\
c & d
\end{bmatrix}
\end{equation*}


$$det(A+B) = det(A) + det(B)$$



\begin{equation*}
\begin{bmatrix}
ta & tb \\
c & d
\end{bmatrix}=
\mathbf{t}\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
\end{equation*}

$$det(tA) = tdet(A) = t^ndet(A)$$

#### 4. If the rows are equal, then $det(A)=0$


#### 5. Subtracting a multiple of one row from another row leaves the same determinant



\begin{equation*}
\begin{bmatrix}
a-lc & b-lc \\
c & d
\end{bmatrix}=
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}+
\begin{bmatrix}
-lc & -ld \\
c & d
\end{bmatrix}
\end{equation*}

맨 오른쪽 식의 판별식은 사라지게 된다

#### 6. If $A$ has a row of zeros, then $det(A)=0$


#### 7. If $A$ is triangular, $det(A)=$ Product of diagnoal entries


#### 8. If $A$ is singular, $det(A)=0$ non-invertible / $A$ is non-singular ( invertible ) $det(A)≠0$


#### 9. $det(AB) = det(A)det(B)$ 

$det(A^{-1}) = \frac{1}{det(A)}$

(1) $B=I$

(2) row exchange of $A$


#### 10. $det(A^T) = det(A) $

$PA = LDU$

$det(PA) = det(P)det(A) = det(L)det(D)det(U)$
    
$det(PA)^T = det(A^T)det(P^T) = det(U^T)det(D^T)det(L^T)$

$=det(L)det(D)det(U)$
