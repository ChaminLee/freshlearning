## Fourier Series 

Orthogonal basis functions ; { $\sin nx , \cos mx$ } n,m : integer

$f(x) = a_0 + \sum_{n=1}^∞ a_n \cos nx + \sum_{m=1}^∞ b_m \sin mx $

$a_1 = \frac{\int f(x) \cos x dx}{(\cos x \cos x)}$

=> Fourier Series is a Projection onto $\cos nx $ and $\sin mx$


## Gram-Schmitt

* Polynomial functions : $ 1, x, x^2 \cdots $ are independent
    ,but not orthogonal for $0≤x≤1$
    
change interval ->  -1 ≤ x ≤ 1

$ a_j - \sum_{i=1}^{j-1} (q_i^Ta_j)q_i \bot q_i ( i=1,2,3,\cdots, j-1 ) $

(1) $b_1(x)=1$

(2) $x - (q_1(x)x) \times 1$ = $x - \int_{-1}^1 x dx \times 1$ =$q_2(x)$

$(b_1(x),b_2(x)) = (1,x) = \int_{-1}^1 x  dx = 0$ -> Orthogonal

(3) $b_3(x) = x^2 - (q_1(x)x^2)q_1(x) - (q_2(x)x^2)q_2(x)$

= $x^2 - \int_{-1}^1 x^2 dx \times 1 -  \int_{-1}^1 x^2 x dx \times x$

= $x^2 - \frac{1}{3}$
