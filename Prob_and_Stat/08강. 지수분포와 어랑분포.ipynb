# 제 08강. 지수분포와 어랑분포

## Poisson Distribution

Random Value X가 시간동안에 베르누이 분포에서 성공한 events의 수를 말한다. 

$P_X(X=k) = \frac{\lambda^k}{K!} e^{-\lambda} , K=0,1,2,\cdots$

$E[X] = \lambda$ 

$\sigma_X^2 = \lambda$

## Exponential Distribution

Random Value X는 lifetime이다. 최대 time까지 살아남을 확률을 말한다. 

pdf : $ f_X(x) = \lambda e^{-\lambda x}$

CDF : $ F_X(x) = \int_0^x \lambda e^{-\lambda t} dt = 1-e^{-\lambda x} = P(X ≤x) $

결국 time x까지 살아남을 확률을 말하는 것이다. 

$E[X] = \frac{1}{\lambda}$

$E[X^2] = \frac{2}{\lambda^2}$

$\sigma_X^2 = \frac{1}{\lambda^2}$

그리고 앞 장의 Geometric distribution 처럼 exponential distribution도 Forgetfulness Property를 가진다.

## Erlang Distribution

E.D (Exponential) 는 time에 대해서, P.D (Poisson)은 events의 수에 대한 분포이다.

Erlang Distribution은 exponential dist의 일반화된 형태이다. 

Random value $X_K$는 시간 간격 K+1의 성공한 events이다.

즉, exponential dist구간을 K개 만큼 더한 것이 Erlang-K가 되는 것이다. 

(K=2면 exp dist에서 2개 집어서 하나로 봄)

${f_X}_{k}(x) = \frac{\lambda^k x^{k-1} e^{-\lambda x}}{(k-1)!}$

$ k = 1,2,3,\cdots / x≥0$일 때,

${F_X}_{k}(x) = P(X_k ≤ x)$

$E[X_k] = \int_0^∞ x \frac{\lambda^k x^{k-1} e^{-\lambda x}}{(k-1)!} dx$

$ = \frac{1}{(k-1)!}  \int_0^∞ (\lambda x)^k  e^{-\lambda x} dx $

$ = \frac{1}{(k-1)!}  \int_0^∞ u^k  e^{-u} \frac{1}{\lambda} du $


Gamma funtion은 다음과 같다.

$\Gamma(k+1) =  \int_0^∞ u^k  e^{-u} du$

$\Gamma(k+1) = k!$

따라서 

$E[X] = \frac{1}{\lambda}\frac{1}{(k-1)!} \Gamma(k+1) = \frac{1}{\lambda}\frac{1}{(k-1)!} k! = \frac{k}{\lambda}$


$E[X_k^2] = \int_0^∞ x^2 \frac{\lambda^k x^{k-1} e^{-\lambda x}}{(k-1)!} dx =\frac{1}{(k-1)!} \int_0^∞  \frac{\lambda^{k+1} x^{k+1} }{\lambda} e^{-\lambda x} dx$

$ = \frac{1}{(k-1)!} \frac{1}{\lambda}  \int_0^∞ (\lambda x)^{k+1} e^{-\lambda x} dx$

$ = \frac{1}{(k-1)!} \frac{1}{\lambda}  \int_0^∞ \frac{1}{\lambda} t^{k+1} e^{-t} dt = \frac{1}{(k-1)!} \frac{1}{\lambda^2}\Gamma(k+2) $

$ = \frac{1}{\lambda^2} k(k+1)$

$\sigma_X^2 = \frac{K(K+1)}{\lambda^2} - \frac{K^2}{\lambda^2} = \frac{K}{\lambda^2}$


## Uniform Distribution

구간 $[a,b]$에 대한 분포이다.

$f_X(x) = \frac{1}{b-a}$ 

조건 : $(a≤x≤b)$

$F_X(x) = \frac{x-a}{b-a}$ 

조건 : $(a≤x≤b)$

$E[X]= \frac{a+b}{2}$ 구간에 대한 평균이기에 그냥 산술적으로 구할 수 있다. 

$E[X^2] = \int_a^b \frac{1}{b-a}x^2 dx = \frac{b^3-a^3}{3(b-a)} = \frac{b^2+ab+a^2}{3}$

$\sigma_X^2 = E[X^2]-E[X]^2 = \frac{b^2+ab+a^2}{3} - \frac{(a+b)^2}{4} = \frac{(b-a)^2}{12}$
